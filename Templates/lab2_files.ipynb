```python
import os
import sys
import re
import nltk
import operator
import pickle as pkl
import time
import numpy as np
import pandas as pd
import nltk
# nltk.download('all')
from nltk.stem import PorterStemmer

def tokenizer(text):
    text = re.sub("[^a-zA-Z]+", " ", text)
    tokens = nltk.tokenize.word_tokenize(text)
    return tokens        

def preprocessing_txt(text):
    
    tokens = tokenizer(text)
    stemmer = nltk.stem.porter.PorterStemmer()
    stopwords = nltk.corpus.stopwords.words('english')
    new_text = ""
    for token in tokens:
        token = token.lower()
        if token not in stopwords:
#             print token
            new_text += stemmer.stem(token)
            new_text += " "
        
    return new_text
```

```python
class BooleanQuery(object):
    def __init__(self, str=None):
        self.str = str
 
    def boolean_and(self):
        str = self.str
        arr = []
        str_ = str.split(" ")
        for strs in str_:
            _ = strs.split(",")
            _ = [" '{}' ".format(x.strip().lower()) for x in _]
            _ = " AND ".join(_)
            _ = "({})".format(_)
            arr.append(_)
        final_query = " OR ".join(arr)
        return final_query
 
 
def main():
    str = 'experimental investigation cake xyz,abc'
    helper = BooleanQuery(str=str)
    query = helper.boolean_and()
    print(query)
 
 
if __name__ == "__main__":
    main()
```
